```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

```python
df['total_rooms'].plot.hist()
```

```python
fig, ax = plt.subplots()
#plot histogram
ax.hist(df['total_bedrooms'])
#set title and labels
ax.set_title("Total Number Of Rooms")
ax.set_xlabel('total_rooms')
ax.set_ylabel('Frequency')
```

```python
df['total_rooms'].plot(kind= 'density', subplots=True, layout=(1,2),
                       sharex=False, figsize=(10,4));
```

```python
# Box plot in seaborn
import seaborn as sns

sns.boxplot(x= 'total_rooms', data=df)
```

```python
# Scatter plot in seaborn
sns.scatterplot(x='total_rooms', y='total_bedrooms', data= df)
```

```python
# Bar plot in seaborn
sns.countplot(x='total_rooms', data= df)
```
