Creating a new column

```python
# Add columns Bedrooms Per Rooms
df["bedrooms_per_rooms"] = df["total_rooms"] / df["total_bedrooms"]
df.head()
```

```python
# Renaming columns
df.rename(
    columns={"total_bedrooms": "BEDROOMS",
             "total_rooms": "ROOMS"}
)
```

```python
# Extracting specific columns
df1 = df["median_house_value"]
df1.head()
```

```python
df2 = df[["total_bedrooms", "total_rooms", "median_house_value"]]
df2.head()
```

Selecting rows

```python
# Selecting rows
df.iloc[9:14] # iloc is used for specifically extracting rows
```

```python
# conditional statements
df1 = df[df["total_rooms"] > 100]
df1.head()
```

```python
# extracting specific rows and specific columns
df.iloc[9:14, 3:5]
```

```python
# conditional statements and column selection
df1 = df.loc[df["total_rooms"] > 100, "median_house_value"]
df1.head()
```
